🧠 What Are Neurons?

So, what exactly are neurons in deep learning?
Neurons are the basic building blocks of deep learning models.
They’re inspired by the biological neurons in our brains. In artificial neural networks (ANNs), neurons—also called nodes—and these are the small computational units that:
- Take in input data
- Perform a mathematical operation
- Pass the result to the next layer of neurons

If that sounds confusing, don’t worry—we’re going to break it down step by step.

🧱 Why Are They Called Building Blocks?

Let’s take a moment to understand why neurons are called the building blocks.

Think of anything around you—images, sounds, even your body.
Everything is made up of tiny units:
- Images are made of pixels
- Humans are made of cells
- Materials are made of atoms

Similarly, deep learning models are made of neurons.
They’re the smallest functional units that, when combined, form powerful models capable of recognizing faces, translating languages, and even driving cars.

🧬 Biological vs Artificial Neurons

Now, you’ve probably heard that artificial neurons are inspired by biological ones.
Let’s explore that.

In your brain:
- A neuron receives signals from other neurons
- It processes those signals
- Then it passes the result to the next neuron

In deep learning:
- An artificial neuron receives numerical inputs
- It applies a mathematical function
- Then it sends the output to the next neuron in the network
So, the inspiration is clear—it’s all about receiving, processing, and passing information.

Imagine a neuron like a little machine with wires coming in from all directions.
Each wire represents an inputs $x_i$ —and there can be as many as you want.

Here’s how a single artificial neuron works:
Step 1: Weighted Sum
Here are these inputs and from this side we give inputs to our neurons
Each input $x_i$ is multiplied by a weight $w_i$.
Then, all these products are summed up:
lets called it z

$z = w_1x_1 + w_2x_2 + \dots + w_nx_n + b$
Here,
- $x_1, x_2, \dots, x_n$ are the inputs
- $w_1, w_2, \dots, w_n$ are the weights
- $b$ is the bias term
- z is the linear combination of inputs


Step 2: Activation Function
This value z is then passed through an activation function to introduce non-linearity:
a = \text{Activation}(z)
There are so many activation functions in the market and but we are not going into the details of actvation function in this video. just think of activation function as some mathematical function.
Some common activation functions are:
- Sigmoid: $\sigma(z) = \frac{1}{1 + e^{-z}}$
- ReLU: $\text{ReLU}(z) = \max(0, z)$
- Tanh: $\tanh(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$
The output a is then passed to the next neuron in the network.

🔄 How Neurons Work Together
In a neural network, neurons are organized into layers:
- Input Layer: Takes in raw data
- Hidden Layers: Perform computations
- Output Layer: Produces the final result
Each neuron in one layer is connected to neurons in the next layer.
This creates a network where data flows forward, and learning happens through backpropagation—a process where the model adjusts weights based on errors.
